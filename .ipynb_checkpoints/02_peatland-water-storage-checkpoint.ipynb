```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random

from functions import *
```

```python
'''Import Data'''
mefSeasonal = pd.read_csv('./mef-data/seasonal/historical_water_storage_processed.csv')

#Sort seasons appropriately
storage_it = mefSeasonal.sort_values(['YEAR', 'Season'], ascending=[True, False])
storage_it.head()
```

```python
'''Distribution plots'''
fig, axs = plt.subplots(5, 4, figsize = (10, 8), 
                    layout = 'tight')

for i in range(0, 20):
    ax = axs.ravel()[i]

    sns.histplot(x = storage_it.iloc[:, i], bins = 15, kde = True, ax = ax)
```

```python
def makeTE_diagram(x, y, lags, title):
    t = []
    tz = []
    sig = []

    for l in lags:
        a, b, c = te_test(x, y, lag = l)
        t.append(a)
        tz.append(b)
        sig.append(c)

    _, ax = plt.subplots(1, 1, figsize = (5, 3))

    sns.lineplot(x = lags, y = t, label = 'TE', ax = ax)
    sns.lineplot(x = lags, y = tz, linestyle = '--', label = r'$TE_z$', ax = ax)

    ax.legend(bbox_to_anchor = (1, 1))
    ax.set_xlim(min(lags), max(lags))
    ax.set_xlabel('Delay, days')
    ax.set_ylabel('Transfer Entropy')
    ax.set_title(title, size = 'small', loc = 'left')
```

```python
#go back 5 years
lags = range(1, 11)
makeTE_diagram(storage_it.South_PCP, storage_it.deltaS, lags, r'Precip -> $\Delta Storage')
```
