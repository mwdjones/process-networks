```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random

from functions import *
```

```python
'''Import Data'''
mefSeasonal = pd.read_csv('./mef-data/seasonal/historical_water_storage_processed.csv')

#Sort seasons appropriately
storage_it = mefSeasonal.sort_values(['YEAR', 'Season'], ascending=[True, False]).reset_index(drop = True)
storage_it.head()
```

```python
#undo storage differentials
storage_it['PAW'] = storage_it.deltaPAW.cumsum()
storage_it['WTE'] = storage_it.deltaWTE.cumsum()
storage_it['S'] = storage_it.deltaS.cumsum()

#Remove first row
storage_it = storage_it.iloc[1:, :]
```

```python
'''Distribution plots'''
fig, axs = plt.subplots(5, 4, figsize = (10, 8), 
                    layout = 'tight')

for i in range(0, 20):
    ax = axs.ravel()[i]
    ax.hist(x = storage_it.iloc[:, i + 3], bins = 15)
    ax.set_title(storage_it.columns[i + 3], size = 'small', loc = 'left')
```

```python
def plot_te(x, y, lags, title, ax):
    t = []

    for l in lags:
        a = te(x, y, lag = l)
        t.append(a)

    sns.lineplot(x = lags, y = t, ax = ax)

    ax.set_xlim(min(lags), max(lags))
    ax.set_xlabel('Delay, days')
    ax.set_ylabel('Transfer Entropy')
    
    
def plot_tetest(x, y, lags, title, ax):
    t = []
    tz = []
    sig = []

    for l in lags:
        a, b, c = te_test(x, y, lag = l)
        t.append(a)
        tz.append(b)
        sig.append(c)

    sns.lineplot(x = lags, y = t, label = 'TE', ax = ax)
    sns.lineplot(x = lags, y = tz, linestyle = '--', label = r'$TE_z$', ax = ax)

    ax.legend(bbox_to_anchor = (1, 1))
    ax.set_xlim(min(lags), max(lags))
    ax.set_xlabel('Delay, seasons')
    ax.set_ylabel('Transfer Entropy')
    ax.set_title(title, size = 'small', loc = 'left')
```

```python
#go back 5 years
lags = range(1, 11)
```

```python
#Make plots
fig, [[ax1, ax2],[ax3, ax4],[ax5, ax6]] = plt.subplots(3, 2, figsize = (7, 8), 
                                                       layout = 'tight', 
                                                       sharey = True)

plot_tetest(storage_it.South_PCP, storage_it.q_interval_cm, lags, r'Precip -> Q', ax = ax1)
plot_tetest(storage_it.South_PCP, storage_it.S, lags, r'Precip -> Storage', ax = ax2)

plot_tetest(storage_it.South_PCP, storage_it.WTE, lags, r'Precip ->  WTE', ax = ax3)
plot_tetest(storage_it.South_PCP, storage_it.PAW, lags, r'Precip -> PAW', ax = ax4)

plot_tetest(storage_it.WTE, storage_it.q_interval_cm, lags, r'WTE -> Q', ax = ax5)
plot_tetest(storage_it.PAW, storage_it.q_interval_cm, lags, r'PAW -> Q', ax = ax6)
```

```python
fig, [[ax1, ax2],[ax3, ax4],[ax5, ax6]] = plt.subplots(3, 2, figsize = (7, 8), 
                                                       layout = 'tight', 
                                                       sharey = True)

plot_te(storage_it.South_PCP, storage_it.q_interval_cm, lags, r'Precip -> Q', ax = ax1)
plot_te(storage_it.South_PCP, storage_it.S, lags, r'Precip -> Storage', ax = ax2)

plot_te(storage_it.South_PCP, storage_it.WTE, lags, r'Precip -> WTE', ax = ax3)
plot_te(storage_it.South_PCP, storage_it.PAW, lags, r'Precip -> Delta', ax = ax4)

plot_te(storage_it.WTE, storage_it.q_interval_cm, lags, r'WTE -> Q', ax = ax5)
plot_te(storage_it.PAW, storage_it.q_interval_cm, lags, r'PAW -> Q', ax = ax6)
```
